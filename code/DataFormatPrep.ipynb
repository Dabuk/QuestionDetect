```python
# CoreNLP POS Tags
annotations = open('path/to/corenlp/.conll').read().split('\n')
```

```python
tempannots = []
posdata = []
j = 0
with open('output.txt','w') as f:
    for i, x in enumerate(annotations):
        if j%5000 == 0:
            print("processed ", j, " lines")
        if x == '':
            zeros = ['0'] * len(data[j].strip().lower().split())
            zeros = ' '.join(zeros)
            f.write(zeros + '\t' + data[j].strip().lower()+'\t'+' '.join(tempannots).strip())
            f.write('\n')
            tempannots = []
            j+=1
            continue
        temp = x.split('\t')
        tempannots.append(temp[3])
f.close()
```
